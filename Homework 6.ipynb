```python
#Question 2
import pandas as pd
url = "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/3_TwoNumOrdered.csv"
df = pd.read_csv(url, delim_whitespace=True)
x = df.iloc[:,0]
y = df.iloc[:,1]
plt.plot(x,y)
plt.xlabel('Date')
plt.ylabel('Value of Bitcoin')
plt.title('Bitcoin market value from April 2013 - April 2018')
plt.show()
```
